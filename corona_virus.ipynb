```python
import requests
import json
import csv
```

```python
response = requests.get('http://coronavirus.m.pipedream.net/')
print(response.text)
json_res = response.content
```

```python
data = json.loads(json_res)
data
```

```python
data['summaryStats']['global']['confirmed']
```

```python
data['rawData'][0]['Country_Region']
data['rawData'][0]['Confirmed']
data['rawData'][0]['Deaths']
```

```python
for country in data['rawData']:
    country_name = country['Country_Region']
    confirmed = country['Confirmed']
    deaths = country['Deaths']
    print(f'name: {country_name}, confirmed: {confirmed}, deaths:{deaths}')
```

```python
with open('covid.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=['Country_Name', 'Deaths', 'Confirmed'])
    writer.writeheader()
    for country in data['rawData']:
        country_name =country['Country_Region']
        confirmed =country['Confirmed']
        deaths =country['Deaths']
        writer.writerow({'Country_Name':country_name, 'Deaths': deaths, 'Confirmed':confirmed})
```
