```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://store.steampowered.com/search/?sort_by=_ASC&filter=topsellers'
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
div = soup.find('div', {'id':'search_result_container'}) 
for a in div.find_all('a', class_='search_result_row'):
    span = a.find('span', class_='title')
    print(span.text)
    print(a['href'])
    print('\n')
```

```python
date = soup.find_all('div', attrs={'class':'col search_released responsive_secondrow'})
for d in date:
    print(d.get_text())
```
